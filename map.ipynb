```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
import numpy as np

plt.figure(figsize=(12, 6))
ax = plt.axes(projection=ccrs.PlateCarree())
ax.set_global()
ax.coastlines()
ax.add_feature(cfeature.BORDERS, linestyle=':')
ax.add_feature(cfeature.LAND, facecolor='lightgray')
ax.add_feature(cfeature.OCEAN, facecolor='lightblue')

# Define rectangle corners
lat_min, lat_max = 18, 47
lon_min, lon_max = 132, 180

# Rectangle polygon coordinates
lons_rect = [lon_min, lon_max, lon_max, lon_min, lon_min]
lats_rect = [lat_min, lat_min, lat_max, lat_max, lat_min]

ax.fill(
    lons_rect, lats_rect,
    color='blue', alpha=0.5,
    transform=ccrs.PlateCarree()
)

plt.title('Area of Interest in the Pacific Ocean')
plt.show()
```
